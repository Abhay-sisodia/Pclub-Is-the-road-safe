```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('accident-data.csv')
df.head(5)
```

Since accident_index, accident_refernce is unique for each accident .Hence these parameters will not help in our model training. So we need to drop these columns.

```python
df.drop(['accident_index','accident_reference'],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
day=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

y=[]
for i in df['day_of_week'].value_counts().sort_index():
    y.append(i)
y
x=[6,5,4,3,2,7,1]
plt.bar(day,y)
```

We can see that most of the accidents happen on Saturdays. We can elaborate this analysis further to find the time of the day when most casualties happen.

```python
# Extracting the month from date
df['date'] = pd.to_datetime(df['date'])
df['date'].dt.to_period('M')
df['date'].dt.strftime('%m/%Y')
df['mm'] = pd.to_datetime(df['date']).dt.month
df['mm']
```

```python
df['hour'] = pd.to_datetime(df['time'], format='%H:%M').dt.hour
c=[]
for i in range(0,24):
    c.append(i)
plt.bar(c,df['hour'].value_counts().sort_index())
```

Most of the accidents happen at evening time around 5:00 PM .

```python
# Since all the accidents occur in 2020 hence we dont need accident_year column.
df.drop(['accident_year','date','time'],axis=1,inplace=True)
```

```python
# Now we need to analyze latitude and longitude of accident prone areas.

import seaborn as sns
cmap = sns.color_palette("Set2", as_cmap=True)
sns.set(rc={'figure.figsize':(12,8)})
sns.scatterplot(data=df, x="longitude", y="latitude", hue = 'number_of_casualties',
                palette = cmap,  sizes=(20, 200))
```

```python
y=df.number_of_casualties
df.drop('number_of_casualties',axis=1,inplace=True)
```

```python
# Before training our model we need to remove null values
null_latitude_index=df[df.latitude.isnull()].index
null_longitude_index=df[df.longitude.isnull()].index
null_latitude_index-null_latitude_index
```

So we have null values in latitude and longitude at the same index position hence we need to fix them.
Since the dataset is large as compared to null values present hence we can randomly impute them with any value (Here we are choosing mean value for imputation).

```python
df['latitude'].fillna(df['latitude'].mean(), inplace = True)
df['longitude'].fillna(df['longitude'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python

dataplot = sns.heatmap(df.corr(), cmap="YlGnBu")
```

```python
#  After analyzing certain key factors we can now proceed to our model training 

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.33, random_state=42)
y_train.to_numpy()
y_test.to_numpy()
```

```python
# from sklearn.tree import DecisionTreeRegressor
# model=DecisionTreeRegressor()

# from sklearn.ensemble import RandomForestRegressor
# model=RandomForestRegressor()
from sklearn.linear_model import LinearRegression
model=LinearRegression()

model.fit(x_train,y_train)
y_pred=model.predict(x_test)

```

```python
type(y_test),type(y_pred)
```

```python

from sklearn.metrics import mean_absolute_error
acc=mean_absolute_error(y_test,y_pred)
acc
```

```python

from sklearn.metrics import mean_squared_error
acc2=mean_squared_error(y_test,y_pred)
acc2

```
